# Are we Exclusive OR not????

We intercepted this encrypted message. Our intelligence suggests it was encrypted using a single-byte XOR with the hex key 0x42. Can you recover the original message?

150a70727077393a72301d2b381d2b2f32723036032c1663633f

```python
# xor each byte of the ct with the key
def xor_decrypt(ct_hex, key):
    ct = bytes.fromhex(ct_hex)
    m = bytes([b ^ key for b in ct])
    return m.decode('utf-8', errors='replace') # exlude errors

ct_hex = "150a70727077393a72301d2b381d2b2f32723036032c1663633f"
key = 0x42
flag = xor_decrypt(ct_hex, key)
print(flag)
```
